```python
#create a function to do create above list
list2 = []
myLabels = []
for gen in movies.Genre.cat.categories:
    list2.append(movies[movies.Genre == gen].AudienceRating)
    myLabels.append(gen)
plt.hist(list2, bins=20, stacked=True, label = myLabels)
plt.legend()
plt.show()
```

Working with FacetGrid

```python
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50,linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
```

```python
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
g = g.map(plt.hist, 'CriticRating')
g.add_legend()
```

```python
#adding co-ordinates and diagonals
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50,linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
g.set(xlim=(0,100),ylim=(0,100))
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='gray', ls='--')
g.add_legend()
```

Dashboards and Styling

```python
sns.set_style("dark", {"axes.facecolor":"black"})
f, axes = plt.subplots(2, 2, figsize=(15,15))

k1 = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, shade=True, shade_lowest=True, cmap='inferno', ax=axes[0,0])
k1b = sns.kdeplot(movies.BudgetMillions, movies.CriticRating, cmap='cool', ax=axes[0,0])

k2 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, shade=True, shade_lowest=True, cmap='inferno', ax=axes[0,1])
k2b = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, cmap='cool', ax=axes[0,1])

z = sns.violinplot(data=movies, x='Year', y='BudgetMillions', ax=axes[1,0], palette='YlOrRd')

k4 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, shade=True, shade_lowest=False,cmap='Blues_r', ax=axes[1,1])
k4b = sns.kdeplot(movies.CriticRating, movies.AudienceRating, cmap='gist_gray_r', ax=axes[1,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

Finalizing the Chapter

```python
#create a function to do create above list
list2 = []
myLabels = []
for gen in movies.Genre.cat.categories:
    list2.append(movies[movies.Genre == gen].AudienceRating)
    myLabels.append(gen)

sns.set_style('whitegrid')
fig,ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)#size of A4 paper
plt.hist(list2, bins=20, stacked=True, label = myLabels)
plt.title("Movie Budget Distribution", fontsize=35, color='DarkBlue', fontname='Times New Roman')
plt.xlabel("Budget", fontsize=25, color='Green')
plt.ylabel("Number of Movies", fontsize=25, color='Red')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size':20})
plt.show()
```
